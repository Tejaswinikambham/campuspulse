```python
#preprocessing
#check null values
print(df.isnull().sum())
```

```python
#checking invalid values
print(df[(df["bedrooms"]<=0) | (df["bedrooms"]<=0)] )
print(df[(df["floors"]<=0) | (df["sqft_lot"]<=0) | (df["sqft_living"]<=0)])
```

```python
#dropping the invalid rows instead of replacing them as the dataset is large removing very few doesnt effect the model much
print( df[(df["bedrooms"]>0) | (df["bedrooms"]>0)] )
```

```python
#id and date doesnt contribute anything to house price so dropping those columns
df = df.drop( ['date','id'], axis=1, errors='ignore')
#analysing how much each feature contributes through correlation
corr_matrix = df.corr()
corr_with_price = corr_matrix['price'].sort_values(ascending=False)
corr_with_price
```

```python
#separating the target and features
y = df['price']
X = df.drop('price', axis=1)
```

```python
#separating the data int 75% and 25% for training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=30)
```

```python
#training the model by giving the data
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#testing the model
y_pred = model.predict(X_test)
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

#checking the performance of the model

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R² Score:", r2)
```
